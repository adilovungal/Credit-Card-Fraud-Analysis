```python
class_amount= df.groupby('Class')['Amount'].agg(['mean','sum','max','min','count']).reset_index()
class_amount
```

```python
df['Hour'] = (df['Time'] // 3600) % 24
pivot_hour = pd.pivot_table(
    df,
    index='Hour',
    columns='Class',
    values='Amount',
    aggfunc=['mean','median','min','max']
)
print(pivot_hour)
```

Distribution of normal and fraudulent transactions

```python
sns.countplot(x='Class', data=df)
plt.title("Distribution of Normal vs Fraudulent Transactions")
plt.show()
```

Average transaction amount by class

```python
#Pivot table with multiple aggregations
pivot_mean = pd.pivot_table(df,
                            index='Class',
                            values='Amount',
                            aggfunc=['mean','min','median','max','sum'])
pivot_mean_reset = pivot_mean['mean'].reset_index()

#Bar plot
plt.figure(figsize=(6,4))
sns.barplot(x='Class', y='Amount', data=pivot_mean_reset, color='purple')
plt.title("Average Transaction Amount by Class")
plt.ylabel("Average Amount")
plt.show()
pivot_mean
```

Transaction Amount vs Class (Log Scale)

```python
plt.figure(figsize=(8,4))
sns.boxplot(x="Class", y="Amount", data=df, palette=['lightblue','purple'])
plt.yscale("log")
plt.title("Transaction Amount vs Class (Log Scale)")
plt.show()
```

```python
sns.histplot(df["V1"], bins=50, kde=True,color="purple")
plt.title("Distribution of V1")
plt.show()
```

Transaction Amount Distribution

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Amount'], bins=50, kde=True, color='purple')
plt.title("Transaction Amount Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Time'], bins=50, kde=True, color='purple')
plt.title("Transaction Time Distribution")
plt.show()
```

Amount vs Time by Class

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='Time', y='Amount', hue='Class', data=df, palette={0:'purple', 1:'black'}, alpha=0.5)
plt.title("Amount vs Time by Class")
plt.show()
```

Histograms for V1-V28

```python
v_features = [f'V{i}' for i in range(1,29)]

plt.figure(figsize=(20,15))
for i, col in enumerate(v_features):
    plt.subplot(7,4,i+1)
    sns.histplot(df[col], bins=50, kde=True, color='purple')
    plt.title(col)
    plt.tight_layout()
plt.show()
```
